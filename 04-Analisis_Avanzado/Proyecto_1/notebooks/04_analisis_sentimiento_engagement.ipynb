## Paso 4: Análisis de Sentimiento e Impacto

Analizaremos cómo el sentimiento percibido del post influye en las métricas de engagement.

```python
import sys
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sys.path.append('../src')
from data_loader import load_engagement_data
from visualization import plot_sentiment_impact

df = load_engagement_data('../data/social_media_engagement1.csv')

# Visualizar impacto del sentimiento
plot_sentiment_impact(df)
```

```python
# Comparación cruzada: Plataforma vs Sentimiento
plt.figure(figsize=(12, 6))
sns.barplot(x='platform', y='likes', hue='sentiment_score', data=df)
plt.title('Likes por Plataforma y Sentimiento')
plt.legend(title='Sentimiento')
plt.show()
```

### Conclusiones de Sentimiento
- Los posts con sentimientos positivos tienden a generar mayor engagement en general.
- Sin embargo, la efectividad del sentimiento varía según la plataforma analizada.
