# Proyecto 1: Análisis de Engagement en Redes Sociales
## Paso 2: Análisis Exploratorio de Datos (EDA)

Exploraremos las distribuciones de likes, comentarios y compartidos por plataforma y tipo de contenido.

```python
import sys
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sys.path.append('../src')
from data_loader import load_engagement_data
from visualization import plot_engagement_by_platform, plot_engagement_correlation

df = load_engagement_data('../data/social_media_engagement1.csv')

# Estadísticas descriptivas
print("Estadísticas descriptivas de métricas numéricas:")
display(df.describe())
```

```python
# Visualización de engagement por plataforma
plot_engagement_by_platform(df)
```

```python
# Distribución por tipo de post
plt.figure(figsize=(10, 6))
sns.countplot(x='post_type', data=df, palette='viridis')
plt.title('Distribución de Tipos de Post')
plt.show()
```

```python
# Correlación entre métricas
plot_engagement_correlation(df)
```

### Hallazgos del EDA
- Existe una correlación positiva entre likes, comentarios y compartidos.
- Algunas plataformas muestran un desempeño superior en likes promedio.
- Los tipos de post (video, image, text, carousel) están distribuidos de forma variada.
