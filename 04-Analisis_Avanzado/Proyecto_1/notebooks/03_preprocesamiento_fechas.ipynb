## Paso 3: Preprocesamiento y Variables Temporales

Convertiremos las fechas a formato datetime y extraeremos información valiosa como la hora del post y el día de la semana.

```python
import sys
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sys.path.append('../src')
from data_loader import load_engagement_data
from feature_engineering import process_dates

df = load_engagement_data('../data/social_media_engagement1.csv')

# Procesar fechas
df = process_dates(df)

print("Nuevas columnas temporales creadas:")
display(df[['post_time', 'hour', 'month', 'day_of_week']].head())
```

```python
# Engagement por hora del día
plt.figure(figsize=(12, 6))
sns.lineplot(x='hour', y='likes', data=df, estimator='mean')
plt.title('Likes promedio por Hora del Día')
plt.grid(True)
plt.show()
```

```python
# Engagement por día de la semana
plt.figure(figsize=(10, 6))
order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
sns.barplot(x='day_of_week', y='likes', data=df, order=order, palette='magma')
plt.title('Likes promedio por Día de la Semana')
plt.xticks(rotation=45)
plt.show()
```

### Insights Temporales
- Se identifican 'horas pico' donde el engagement tiende a ser mayor.
- Los fines de semana vs días laborales muestran comportamientos distintos en la interacción de los usuarios.
