## 1. Importación de Librerías

```python
# Librerías de manipulación de datos
import pandas as pd
import numpy as np

# Librerías de visualización
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración de visualizaciones
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)
plt.rcParams['font.size'] = 10

# Suprimir advertencias
import warnings
warnings.filterwarnings('ignore')

print("✓ Librerías importadas correctamente")
```

## 2. Carga del Dataset

El dataset contiene información de clientes mayoristas con variables de gasto en diferentes categorías de productos:
- **Channel**: Tipo de cliente (1: Hotel/Restaurante/Café, 2: Retail)
- **Region**: Región geográfica
- **Fresh**: Gasto en productos frescos
- **Milk**: Gasto en productos lácteos
- **Grocery**: Gasto en comestibles
- **Frozen**: Gasto en productos congelados
- **Detergents_Paper**: Gasto en detergentes y papel
- **Delicassen**: Gasto en delicatessen

```python
# Cargar el dataset
ruta_datos = '../data/Wholesale customers data.csv'
df = pd.read_csv(ruta_datos)

print(f"✓ Dataset cargado exitosamente")
print(f"\nDimensiones: {df.shape[0]} filas × {df.shape[1]} columnas")
```

## 3. Exploración Inicial de la Estructura

```python
# Ver las primeras filas del dataset
print("Primeras 10 filas del dataset:")
print(df.head(10))
```

```python
# Información detallada del dataset
print("\n" + "="*60)
print("INFORMACIÓN DEL DATASET")
print("="*60)
print(df.info())
```

```python
# Estadísticas descriptivas
print("\n" + "="*60)
print("ESTADÍSTICAS DESCRIPTIVAS")
print("="*60)
df.describe().round(2)
```

## 4. Análisis de Valores Faltantes

```python
# Verificar valores faltantes
print("Valores faltantes por columna:")
valores_faltantes = df.isnull().sum()
print(valores_faltantes)

print(f"\nTotal de valores faltantes: {df.isnull().sum().sum()}")

if df.isnull().sum().sum() == 0:
    print("✓ No hay valores faltantes en el dataset")
```

## 5. Análisis de Valores Duplicados

```python
# Verificar duplicados
num_duplicados = df.duplicated().sum()
print(f"Número de filas duplicadas: {num_duplicados}")

if num_duplicados > 0:
    print(f"\nPorcentaje de duplicados: {(num_duplicados/len(df)*100):.2f}%")
```

## 6. Descripción de Variables Categóricas

```python
# Análisis de la variable Channel
print("\nDistribución de Channel (Tipo de Cliente):")
print(df['Channel'].value_counts().sort_index())
print(f"\nPorcentajes:")
print(df['Channel'].value_counts(normalize=True).sort_index() * 100)

# Análisis de la variable Region
print("\n" + "="*60)
print("Distribución de Region:")
print(df['Region'].value_counts().sort_index())
print(f"\nPorcentajes:")
print(df['Region'].value_counts(normalize=True).sort_index() * 100)
```

## 7. Visualización Inicial de Distribuciones

```python
# Visualización de distribuciones por tipo de cliente
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Gráfico de barras para Channel
channel_counts = df['Channel'].value_counts().sort_index()
axes[0].bar(['Hotel/Rest/Café', 'Retail'], channel_counts.values, color=['#2E86AB', '#A23B72'])
axes[0].set_title('Distribución de Tipos de Cliente', fontsize=12, fontweight='bold')
axes[0].set_ylabel('Cantidad de Clientes')
for i, v in enumerate(channel_counts.values):
    axes[0].text(i, v + 10, str(v), ha='center', va='bottom', fontweight='bold')

# Gráfico de barras para Region
region_counts = df['Region'].value_counts().sort_index()
axes[1].bar(region_counts.index, region_counts.values, color='#F18F01')
axes[1].set_title('Distribución de Regiones', fontsize=12, fontweight='bold')
axes[1].set_xlabel('Región')
axes[1].set_ylabel('Cantidad de Clientes')
for i, v in enumerate(region_counts.values):
    axes[1].text(i, v + 5, str(v), ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.show()

print("✓ Gráficos de distribuciones generados exitosamente")
```

## 8. Resumen de Características del Dataset

```python
# Crear un resumen informativo
print("\n" + "="*70)
print("RESUMEN DEL DATASET - CLIENTES MAYORISTAS")
print("="*70)

print(f"\n📊 DIMENSIONES:")
print(f"  • Total de registros: {df.shape[0]}")
print(f"  • Total de variables: {df.shape[1]}")

print(f"\n✓ CALIDAD DE DATOS:")
print(f"  • Valores faltantes: {df.isnull().sum().sum()} (Excelente)")
print(f"  • Filas duplicadas: {df.duplicated().sum()}")

print(f"\n📋 VARIABLES CATEGÓRICAS:")
print(f"  • Channel: {df['Channel'].nunique()} tipos")
print(f"  • Region: {df['Region'].nunique()} regiones")

print(f"\n💰 VARIABLES NUMÉRICAS (Gastos en Unidades Monetarias):")
categorias = ['Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper', 'Delicassen']
for cat in categorias:
    print(f"  • {cat}: ${df[cat].min():,.0f} - ${df[cat].max():,.0f} (Promedio: ${df[cat].mean():,.0f})")

print(f"\n" + "="*70)
```
