## 1. Preparación de Datos y Carga de Resultados

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score, silhouette_samples
from sklearn.decomposition import PCA
import warnings
warnings.filterwarnings('ignore')

# Configuración visual
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (14, 8)

print("✓ Librerías importadas")
```

```python
# Cargar y preparar datos
ruta_datos = '../data/Wholesale customers data.csv'
df_original = pd.read_csv(ruta_datos)

variables_numericas = ['Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper', 'Delicassen']
scaler = StandardScaler()
df_estandarizado = df_original.copy()
df_estandarizado[variables_numericas] = scaler.fit_transform(df_original[variables_numericas])

X = df_estandarizado[variables_numericas].values

# Aplicar K-means final (k=3)
kmeans_final = KMeans(n_clusters=3, random_state=42, n_init=10)
labels_final = kmeans_final.fit_predict(X)

df_original['Cluster'] = labels_final
df_estandarizado['Cluster'] = labels_final

print(f"✓ Datos preparados y K-means (k=3) aplicado")
print(f"Distribución final: {np.bincount(labels_final)}")
```
