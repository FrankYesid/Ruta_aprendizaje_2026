# 01 - Ingesta y perfilado del dataset financiero

En este cuaderno se carga el archivo `financial_timeseries_dataset.csv` desde la carpeta `data`, se explora su estructura y se genera un informe automático de perfilado con `ydata-profiling`.

```python
from pathlib import Path

import pandas as pd
```

## Carga del dataset desde `data/financial_timeseries_dataset.csv`

```python
csv_path = Path("..") / "data" / "financial_timeseries_dataset.csv"
df = pd.read_csv(csv_path)

df.head()
```

## Estructura general del dataset

```python
df.info()
```

```python
df.describe(include="all")
```

## Generación de informe con ydata-profiling

```python
from ydata_profiling import ProfileReport

profile = ProfileReport(df, title="Financial Time Series Profiling", explorative=True)

output_path = Path("..") / "data" / "financial_timeseries_profile.html"
profile.to_file(output_path)
output_path
```
