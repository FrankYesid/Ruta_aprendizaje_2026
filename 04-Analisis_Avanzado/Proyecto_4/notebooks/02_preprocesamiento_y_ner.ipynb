# 02 - Preprocesamiento y Reconocimiento de Entidades Nombradas (NER)
Funciones para limpiar texto y aplicación de un modelo `transformers` de NER para extraer entidades como PERSON, ORG y LOC de los documentos legales.

```python
# Ejemplo de funciones de limpieza
import re

def limpiar_texto(texto):
    texto = texto.lower()
    texto = re.sub(r"\s+", " ", texto)
    texto = re.sub(r"[^a-z0-9áéíóúüñ ]", "", texto)
    return texto.strip()

# Prueba de limpieza
print(limpiar_texto("Este es un Ejemplo, con puntuación! 2026."))
```

```python
# Usar pipeline de transformers para NER
from transformers import pipeline

ner = pipeline("ner", grouped_entities=True)

# ejemplo sencillo
ejemplo = "El contrato fue firmado por Juan Pérez en Madrid." 
print(ner(ejemplo))
```

```python
# integrado con dataset streaming
from datasets import load_dataset
config = "de_caselaw"
dataset = load_dataset("joelito/eurlex_resources", config, split='train', streaming=True)

for i, item in enumerate(dataset):
    if i >= 3:
        break
    text = limpiar_texto(item.get('text',''))
    entidades = ner(text[:1000])  # acota longitud para la demo
    print(f"Documento {i}", entidades)
```
