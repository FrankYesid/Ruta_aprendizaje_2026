# 05 - Búsqueda y similitud semántica
Este cuaderno emplea `sentence-transformers` para convertir los textos en embeddings y realizar búsquedas mediante similitud de coseno.

```python
from datasets import load_dataset
import itertools

config = "de_caselaw"
dataset = load_dataset("joelito/eurlex_resources", config, split='train', streaming=True)

# cargar algunos textos
texts = [item.get('text','') for item in itertools.islice(dataset, 500)]

from sentence_transformers import SentenceTransformer
import numpy as np

model = SentenceTransformer('all-MiniLM-L6-v2')
embeddings = model.encode(texts, show_progress_bar=True)

# función de búsqueda
def buscar(query, top_k=5):
    q_vec = model.encode([query])[0]
    sims = np.dot(embeddings, q_vec) / (np.linalg.norm(embeddings, axis=1) * np.linalg.norm(q_vec))
    idxs = np.argsort(sims)[-top_k:][::-1]
    for i in idxs:
        print(f"Score {sims[i]:.4f}: {texts[i][:500]}\n")

# ejemplo
encontrar = buscar("contrato de arrendamiento")
```
