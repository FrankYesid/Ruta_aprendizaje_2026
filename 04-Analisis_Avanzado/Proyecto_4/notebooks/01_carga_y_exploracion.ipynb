# 01 - Carga y exploración del dataset eurlex_resources
Este cuaderno muestra cómo cargar el dataset **`eurlex_resources`** desde Hugging Face en modo *streaming* para manejar su gran tamaño. Se incluyen ejemplos de iteración, visualización de textos y análisis básico de la longitud de los documentos.

```python
from datasets import load_dataset
config = "de_caselaw"  # {lang}_{resource}
dataset = load_dataset("joelito/eurlex_resources", config, split='train', streaming=True)

# Iterar y mostrar algunos ejemplos
for i, example in enumerate(dataset):
    if i >= 5:
        break
    print(f"--- Ejemplo {i+1} ---")
    print(example.get('text', '')[:1000])
    print()
```

```python
# analizar la distribución de longitudes (muestrear primeros 1000 documentos)
import itertools
import numpy as np

lengths = []
for i, example in enumerate(itertools.islice(dataset, 1000)):
    text = example.get('text', '')
    lengths.append(len(text.split()))

import matplotlib.pyplot as plt
plt.hist(lengths, bins=30);
plt.title('Distribución de longitudes (palabras)');
```
