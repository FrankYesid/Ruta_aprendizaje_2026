# Proyecto 4: Machine Learning Explicable (XAI) - LIME

LIME (Local Interpretable Model-agnostic Explanations) es una técnica que explica las predicciones de cualquier clasificador o regresor de aprendizaje automático de forma local, aproximándolo con un modelo interpretable.

## 1. Librerías y Carga de Modelo/Datos

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import joblib
import lime
import lime.lime_tabular

# Cargar modelo y datos
model = joblib.load('../models/random_forest_boston.pkl')
X_train = pd.read_csv('../data/X_train.csv')
X_test = pd.read_csv('../data/X_test.csv')
y_test = pd.read_csv('../data/y_test.csv')

print("Modelo y datos cargados.")
```

## 2. Creación del Explicador LIME

LIME requiere el conjunto de entrenamiento para entender la distribución de los datos.

```python
explainer = lime.lime_tabular.LimeTabularExplainer(
    training_data=np.array(X_train),
    feature_names=X_train.columns,
    class_names=['price'],
    mode='regression'
)

print("Explicador LIME inicializado.")
```

## 3. Explicación de una Predicción Local

Elegimos la misma observación que en el cuaderno de SHAP para comparar.

```python
sample_idx = 0
exp = explainer.explain_instance(
    data_row=X_test.iloc[sample_idx],
    predict_fn=model.predict
)

print(f"Predicción LIME: {exp.predicted_value:.2f}")
print(f"Valor real: {y_test.iloc[sample_idx, 0]:.2f}")
```

### 3.1 Visualización de la Explicación

LIME muestra qué características contribuyen a aumentar o disminuir el valor predicho localmente.

```python
# Visualización en el cuaderno
exp.show_in_notebook(show_table=True)
```

### 3.2 Visualización Estática (Gráfico de Barras)

```python
exp.as_pyplot_figure()
plt.title('Explicación LIME Local')
plt.show()
```

## 4. Comparación SHAP vs LIME

- **SHAP**: Basado en fundamentos teóricos sólidos (valores de Shapley), garantiza consistencia y aditividad.
- **LIME**: Más rápido y flexible, pero puede ser inestable (las explicaciones pueden variar ligeramente entre ejecuciones debido a su naturaleza de muestreo local).

Ambos coinciden en que `rm` y `lstat` son críticos para este dataset.
