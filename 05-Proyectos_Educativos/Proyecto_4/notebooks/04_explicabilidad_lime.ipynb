# Proyecto 4: Machine Learning Explicable (XAI) - LIME

LIME (Local Interpretable Model-agnostic Explanations) explica las predicciones de cualquier clasificador o regresor de caja negra aproximándolo localmente con un modelo interpretable.

## 1. Librerías y Carga de Modelo/Datos

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import joblib
import lime
import lime.lime_tabular

# Cargar modelo y datos
model = joblib.load('../models/random_forest_boston.pkl')
X_train = pd.read_csv('../data/X_train.csv')
X_test = pd.read_csv('../data/X_test.csv')
y_test = pd.read_csv('../data/y_test.csv')

print("Modelo y datos cargados.")
```

## 2. Creación del Explicador LIME

Necesitamos pasar los datos de entrenamiento para que LIME conozca la distribución de las variables.

```python
explainer = lime.lime_tabular.LimeTabularExplainer(
    training_data=np.array(X_train),
    feature_names=X_train.columns,
    class_names=['medv'],
    mode='regression'
)

print("Explicador LIME creado.")
```

## 3. Explicación Local (Una predicción específica)

### 3.1 Elegir una instancia

Vamos a explicar la primera observación del conjunto de prueba.

```python
idx = 0
sample_data = X_test.iloc[idx, :].values
y_real = y_test.iloc[idx, 0]

print(f"Predicción: {model.predict([sample_data])[0]:.2f}")
print(f"Valor Real: {y_real:.2f}")
```

### 3.2 Generar Explicación con LIME

```python
exp = explainer.explain_instance(
    data_row=sample_data,
    predict_fn=model.predict,
    num_features=10
)

exp.show_in_notebook(show_table=True)
```

### 3.3 Visualización de Gráfico LIME

```python
exp.as_pyplot_figure()
plt.title('Explicación LIME para la instancia elegida')
plt.show()
```

## 4. Análisis de Casos Extremos

### 4.1 Predicción Alta

```python
high_idx = y_test['medv'].idxmax()
sample_high = X_test.iloc[high_idx, :].values

exp_high = explainer.explain_instance(sample_high, model.predict, num_features=5)
exp_high.as_pyplot_figure()
plt.title(f'Predicción Alta (Valor Real: {y_test.iloc[high_idx, 0]:.2f})')
plt.show()
```

### 4.2 Predicción Baja

```python
low_idx = y_test['medv'].idxmin()
sample_low = X_test.iloc[low_idx, :].values

exp_low = explainer.explain_instance(sample_low, model.predict, num_features=5)
exp_low.as_pyplot_figure()
plt.title(f'Predicción Baja (Valor Real: {y_test.iloc[low_idx, 0]:.2f})')
plt.show()
```

## 5. Conclusión LIME

- LIME ofrece una visión muy clara de qué variables pesaron más en una decisión específica.
- Es especialmente útil para depurar predicciones individuales.
- A diferencia de SHAP, LIME es una aproximación local rápida y fácil de interpretar visualmente.
