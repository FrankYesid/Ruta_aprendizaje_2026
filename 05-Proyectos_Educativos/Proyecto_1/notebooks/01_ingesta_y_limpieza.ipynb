# 01 - Ingesta y Limpieza de Datos COVID-19

En este cuaderno se descargan las series temporales mundiales de casos confirmados de COVID-19 desde el repositorio oficial de la Universidad Johns Hopkins. Se realiza el preprocesamiento necesario para transformar el formato ancho a largo y preparar los datos para el análisis.

```python
import pandas as pd
import numpy as np
from pathlib import Path

# URLs de los datasets oficiales de Johns Hopkins (GitHub)
BASE_URL = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/"
CONFIRMED_URL = BASE_URL + "time_series_covid19_confirmed_global.csv"
DEATHS_URL = BASE_URL + "time_series_covid19_deaths_global.csv"
RECOVERED_URL = BASE_URL + "time_series_covid19_recovered_global.csv"
```

## 1. Descarga de Datos

```python
print("Descargando datos confirmados...")
df_confirmed = pd.read_csv(CONFIRMED_URL)
print(f"Datos confirmados cargados: {df_confirmed.shape}")

df_confirmed.head()
```

## 2. Transformación de Formato (Melt)

Los datos originales vienen con las fechas en columnas. Necesitamos transformarlos a un formato de serie temporal largo.

```python
def transform_df(df, name):
    df_melted = df.melt(
        id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'], 
        var_name='Date', 
        value_name=name
    )
    df_melted['Date'] = pd.to_datetime(df_melted['Date'])
    return df_melted

df_long = transform_df(df_confirmed, 'Confirmed')
df_long.head()
```

## 3. Guardado en Local

Guardamos los datos procesados en la carpeta `data` para los siguientes cuadernos.

```python
data_dir = Path("..") / "data"
data_dir.mkdir(exist_ok=True)

df_long.to_csv(data_dir / "covid_confirmed_long.csv", index=False)
print(f"Archivo guardado en: {data_dir / 'covid_confirmed_long.csv'}")
```
