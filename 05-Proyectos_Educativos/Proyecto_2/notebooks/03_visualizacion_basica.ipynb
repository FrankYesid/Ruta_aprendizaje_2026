# 03 - Visualización Básica de Datos Limpios

Una vez que el dataset de giras musicales ha sido limpiado y normalizado, el siguiente paso es realizar una exploración visual para entender las distribuciones de los datos y las estadísticas principales de los artistas y sus giras.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

sns.set_style("whitegrid")

# Carga del dataset limpio
data_path = Path("..") / "data" / "music_tours_cleaned.csv"
df = pd.read_csv(data_path)

print(f"Dimensiones del dataset limpio: {df.shape}")
df.head()
```

## 1. Distribución de Ingresos

Analicemos cómo se distribuyen los ingresos reales (`Actual gross`) entre las giras más exitosas.

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Actual gross'], bins=20, kde=True, color='skyblue')
plt.title('Distribución de Ingresos Actuales (en dólares)')
plt.xlabel('Ingresos ($)')
plt.ylabel('Frecuencia')
plt.show()
```

## 2. Comparación de Shows por Artista

Veamos qué artistas tienen más registros (giras) en el Top histórico del dataset.

```python
plt.figure(figsize=(12, 6))
artist_counts = df['Artist'].value_counts()
sns.barplot(x=artist_counts.values, y=artist_counts.index, palette='viridis')
plt.title('Número de Giras en el Top por Artista')
plt.xlabel('Cantidad de Giras')
plt.ylabel('Artista')
plt.show()
```

## 3. Promedio de Ingresos por Show

Analicemos la eficiencia de las giras comparando el promedio de ingresos por cada show realizado.

```python
plt.figure(figsize=(10, 8))
df_sorted = df.sort_values('Average gross', ascending=False).head(15)
sns.barplot(data=df_sorted, x='Average gross', y='Tour title', hue='Artist', dodge=False)
plt.title('Top 15 Giras por Promedio de Ingresos por Show')
plt.xlabel('Promedio por Show ($)')
plt.ylabel('Gira')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```
