# 04 - Análisis Visual Avanzado y Correlaciones

En este cuaderno final de visualización, exploraremos relaciones más complejas entre las variables del dataset de giras musicales, como la correlación entre el número de shows y los ingresos ajustados, y cómo el ranking influye en las métricas económicas.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

sns.set_style("whitegrid")

# Carga del dataset limpio
data_path = Path("..") / "data" / "music_tours_cleaned.csv"
df = pd.read_csv(data_path)

df.head()
```

## 1. Relación entre Shows e Ingresos Ajustados

¿Hacer más shows garantiza mayores ingresos ajustados a la inflación? Analicemos esta relación mediante un gráfico de dispersión.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Shows', y='Adjusted gross (in 2022 dollars)', size='Average gross', hue='Artist', alpha=0.7)
plt.title('Relación: Número de Shows vs Ingresos Ajustados')
plt.xlabel('Número de Shows')
plt.ylabel('Ingresos Ajustados ($)')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

## 2. Matriz de Correlación

Identifiquemos qué variables numéricas tienen una relación más fuerte entre sí.

```python
plt.figure(figsize=(8, 6))
numeric_df = df.select_dtypes(include=[np.number])
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Matriz de Correlación de Variables Numéricas')
plt.show()
```

## 3. Análisis de Ingresos por Posición Pico (Peak)

Veamos si las giras que alcanzaron el puesto #1 tienen una diferencia significativa en ingresos promedio respecto a las demás.

```python
plt.figure(figsize=(10, 6))
df['Is Peak 1'] = df['Peak'] == 1
sns.boxplot(data=df, x='Is Peak 1', y='Average gross', palette='Set2')
plt.title('Distribución de Ingresos Promedio: Giras #1 vs Resto')
plt.xlabel('¿Alcanzó el Puesto #1?')
plt.ylabel('Promedio por Show ($)')
plt.xticks([0, 1], ['No', 'Sí'])
plt.show()
```

## 4. Resumen de Hallazgos

- Existe una correlación positiva moderada entre el número de shows y los ingresos totales.
- Los artistas con mayor ingresos por show no siempre son los que más shows realizan (ej. Taylor Swift).
- El ajuste por inflación permite comparar giras de diferentes décadas de manera más justa.
