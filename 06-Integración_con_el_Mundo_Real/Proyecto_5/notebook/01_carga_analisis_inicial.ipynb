# 01. Carga y análisis inicial

Este cuaderno revisa la estructura del dataset, la calidad de los datos, la distribución de variables y los patrones iniciales para entender el problema de retrasos del transporte público.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

plt.style.use('seaborn-v0_8-darkgrid')

df = pd.read_csv(Path('data/public_transport_delays.csv'))
df.head()
```

## Vista general del dataset

En esta sección se revisa el tamaño del conjunto, los tipos de datos y la presencia de valores faltantes.

```python
print('Filas:', df.shape[0])
print('Columnas:', df.shape[1])
print('\nTipos de datos:')
print(df.dtypes)
print('\nValores faltantes:')
print(df.isna().sum())
```

## Distribución de la variable objetivo

La columna delayed representa si el viaje terminó retrasado o no. Este balance es clave para interpretar los resultados del modelo.

```python
target_dist = df['delayed'].value_counts().rename(index={1:'Retrasado', 0:'No retrasado'})
ax = target_dist.plot(kind='bar', color=['#4C78A8', '#F58518'], edgecolor='black')
ax.set_title('Distribución de la variable delayed')
ax.set_ylabel('Frecuencia')
ax.set_xlabel('Clase')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

## Distribuciones numéricas

Se inspeccionan las distribuciones de las variables numéricas para detectar sesgos, valores extremos o comportamientos particulares.

```python
num_cols = ['actual_departure_delay_min', 'actual_arrival_delay_min', 'temperature_C', 'humidity_percent', 'wind_speed_kmh', 'precipitation_mm', 'event_attendance_est', 'traffic_congestion_index']
fig, axes = plt.subplots(2, 4, figsize=(16, 8))
axes = axes.flatten()
for ax, col in zip(axes, num_cols):
    sns.histplot(df[col], bins=25, kde=True, ax=ax, color='#4C78A8')
    ax.set_title(col)
fig.tight_layout()
plt.show()
```

## Análisis exploratorio por categorías

Se comparan las proporciones de retrasos según transporte, clima, eventos y horarios.

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 8))

sns.countplot(data=df, x='transport_type', hue='delayed', ax=axes[0, 0], palette=['#4C78A8', '#F58518'])
axes[0, 0].set_title('Retrasos por tipo de transporte')

sns.countplot(data=df, x='weather_condition', hue='delayed', ax=axes[0, 1], palette=['#4C78A8', '#F58518'])
axes[0, 1].set_title('Retrasos por condición climática')
axes[0, 1].tick_params(axis='x', rotation=45)

sns.countplot(data=df, x='event_type', hue='delayed', ax=axes[1, 0], palette=['#4C78A8', '#F58518'])
axes[1, 0].set_title('Retrasos por tipo de evento')
axes[1, 0].tick_params(axis='x', rotation=45)

sns.countplot(data=df, x='peak_hour', hue='delayed', ax=axes[1, 1], palette=['#4C78A8', '#F58518'])
axes[1, 1].set_title('Retrasos por hora pico')

fig.tight_layout()
plt.show()
```

### Resumen del notebook 1

Este primer análisis confirma que el problema es de clasificación binaria con un desbalance notable, además de mostrar que las variables de contexto y de servicio parecen estar asociadas con la ocurrencia de retrasos.
