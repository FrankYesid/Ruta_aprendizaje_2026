# 02 EDA y Visualización

Distribuciones de ratings, actividad de usuarios y popularidad de películas.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from pathlib import Path
base = Path(r'd:\GitHub\Ruta_aprendizaje_2024\03-aplicación_de_modelos\proyecto_3\data')
ratings = pd.read_csv(base / 'ratings.csv')
movies = pd.read_csv(base / 'movies.csv')
```

## Distribución de ratings

```python
sns.set(style='whitegrid')
plt.figure(figsize=(8,4))
sns.histplot(ratings['rating'], bins=20, kde=True)
plt.title('Distribución de ratings')
plt.show()
```

## Actividad por usuario y popularidad por película

```python
user_activity = ratings.groupby('userId')['movieId'].count()
movie_popularity = ratings.groupby('movieId')['userId'].count()
plt.figure(figsize=(8,4))
sns.histplot(user_activity, bins=30)
plt.title('Actividad de usuarios (número de ratings por usuario)')
plt.show()
plt.figure(figsize=(8,4))
sns.histplot(movie_popularity, bins=30)
plt.title('Popularidad de películas (número de ratings por película)')
plt.show()
```
