# 01 Ingesta y Estandarización de MovieLens

Carga de datasets, verificación de esquemas y estandarización de tipos y columnas.

```python
import pandas as pd
from pathlib import Path
base = Path(r'd:\GitHub\Ruta_aprendizaje_2024\03-aplicación_de_modelos\proyecto_3\data')
ratings_path = base / 'ratings.csv'
movies_path = base / 'movies.csv'
tags_path = base / 'tags.csv'
links_path = base / 'links.csv'
```

```python
ratings = pd.read_csv(ratings_path)
movies = pd.read_csv(movies_path)
tags = pd.read_csv(tags_path)
links = pd.read_csv(links_path)
ratings.head(), movies.head(), tags.head(), links.head()
```

## Estandarización de tipos y nombres

Conversión de `timestamp` a datetime, normalización de `rating` y separación de géneros.

```python
ratings['timestamp'] = pd.to_datetime(ratings['timestamp'], unit='s', errors='coerce')
ratings['rating'] = ratings['rating'].astype(float)
movies['genres'] = movies['genres'].replace('(no genres listed)', '').str.split('|')
ratings = ratings.drop_duplicates()
movies = movies.drop_duplicates()
tags = tags.drop_duplicates()
links = links.drop_duplicates()
ratings.info(), movies.info(), tags.info(), links.info()
```

## Uniones básicas

Unión de `ratings` con `movies` y `links` para un esquema integrado.

```python
df = ratings.merge(movies, on='movieId', how='left').merge(links, on='movieId', how='left')
df.head()
```
