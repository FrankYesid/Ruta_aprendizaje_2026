# 02 EDA y Visualización (VRP)

Exploración de demanda y distribución espacial de clientes.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from pathlib import Path
base = Path(r'd:\GitHub\Ruta_aprendizaje_2024\03-aplicación_de_modelos\proyecto_2\data')
df = pd.read_csv(base / 'VRP.csv')
df.columns = df.columns.str.lower()
```

## Distribución de demanda

```python
if 'demand' in df.columns:
    plt.figure(figsize=(8,4))
    sns.histplot(df['demand'], bins=30)
    plt.title('Distribución de demanda')
    plt.show()
```

## Mapa de clientes (si hay coordenadas x/y)

```python
if {'x','y'}.issubset(set(df.columns)):
    plt.figure(figsize=(6,6))
    plt.scatter(df['x'], df['y'], s=10)
    plt.title('Distribución espacial de clientes')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()
```
