Carga VRP.csv, verificación de esquema y construcción de insumos.

```python
import pandas as pd
from pathlib import Path
base = Path(r'd:\GitHub\Ruta_aprendizaje_2024\03-aplicación_de_modelos\proyecto_2\data')
df = pd.read_csv(base / 'VRP.csv')
df.head()
```

## Preparación de demanda y coordenadas

```python
cols = df.columns.str.lower()
df.columns = cols
if 'demand' in df.columns:
    df['demand'] = df['demand'].astype(float)
if 'x' in df.columns and 'y' in df.columns:
    df['x'] = df['x'].astype(float)
    df['y'] = df['y'].astype(float)
df.describe(include='all')
```
