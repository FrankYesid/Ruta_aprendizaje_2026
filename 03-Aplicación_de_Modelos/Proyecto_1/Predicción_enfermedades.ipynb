```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1.1 Carga de librería de pandas y dataset
# Reemplaza 'dataset.csv' con el nombre real del archivo
file_path = "dataset.csv"
df = pd.read_csv(file_path)

# 1.2 Número de variables y observaciones del dataset
num_filas, num_columnas = df.shape
print(f"El dataset tiene {num_filas} filas y {num_columnas} columnas.")

# 1.3 Nombre de las variables
print("Las variables del dataset son:")
print(df.columns.tolist())

# 1.4 Características generales de las variables
df_info = df.info()

# 1.5 Visualización de las primeras filas del dataset
print("Primeras filas del dataset:")
print(df.head())

# 1.6 Datos estadísticos generales del dataset
print("Descripción estadística del dataset:")
print(df.describe())

# 1.7 Correlación entre las variables numéricas
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Matriz de correlación")
plt.show()
```
