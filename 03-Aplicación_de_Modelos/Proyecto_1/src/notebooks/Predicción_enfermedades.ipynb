### BRFSS Dataset

El **Behavioral Risk Factor Surveillance System (BRFSS)** es el sistema líder en encuestas telefónicas relacionadas con la salud en EE.UU. Recolecta datos estatales sobre los comportamientos de riesgo relacionados con la salud, condiciones crónicas y uso de servicios preventivos.

Este conjunto de datos proporciona información clave para la predicción de enfermedades cardiovasculares basándose en factores de estilo de vida.

El conjunto de datos original contenía **304 variables únicas**, de las cuales se seleccionaron **19** relacionadas con factores de estilo de vida que pueden contribuir al riesgo de desarrollar alguna enfermedad cardiovascular. Se aplicaron técnicas de limpieza y balanceo de clases para mejorar la calidad de los datos.

````python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el dataset
file_path = 'cardiovascular_diseases_risk.csv'  # Asegúrate de tener el archivo en el directorio correcto
df = pd.read_csv(file_path)

# 2.1 Comprobación de la existencia de datos NA
print("Valores nulos por columna:")
print(df.isnull().sum())

# 2.2 Comprobación de outliers
plt.figure(figsize=(12, 6))
sns.boxplot(data=df.select_dtypes(include=[np.number]))
plt.xticks(rotation=90)
plt.title("Detección de outliers")
plt.show()

# 2.3 Crear columna con total de enfermedades
enfermedades = ['Heart_Disease', 'Skin_Cancer', 'Other_Cancer', 'Depression', 'Diabetes', 'Arthritis']
df['Total_Diseases'] = df[enfermedades].apply(lambda x: x.eq('Yes').sum(), axis=1)

# 2.4 Crear columna con el estado de peso en función del BMI
# Suponiendo que la columna 'BMI' existe en el dataset
def clasificar_peso(bmi):
    if bmi < 18.5:
        return 'Bajo peso'
    elif 18.5 <= bmi < 24.9:
        return 'Normal'
    elif 25 <= bmi < 29.9:
        return 'Sobrepeso'
    else:
        return 'Obesidad'

if 'BMI' in df.columns:
    df['Weight_Status'] = df['BMI'].apply(clasificar_peso)

# 2.5 Ordenar columnas
columnas_ordenadas = ['Sex', 'General_Health', 'Checkup', 'Exercise', 'Weight_Status', 'Total_Diseases'] + enfermedades
df = df[columnas_ordenadas]

# 2.6 Cambiar dtypes
categoricas = ['Sex', 'General_Health', 'Checkup', 'Exercise', 'Weight_Status']
for col in categoricas:
    df[col] = df[col].astype('category')

# 2.7 Normalización de ciertas variables
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

if 'BMI' in df.columns:
    df['BMI'] = scaler.fit_transform(df[['BMI']])

# Guardar el dataset transformado
df.to_csv('processed_cardiovascular_diseases_risk.csv', index=False)

print("Preprocesamiento completado.")
````

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el dataset
df = pd.read_csv("cardiovascular_diseases_risk.csv")

# 3.1 Número de hombres y mujeres
gender_counts = df['Sex'].value_counts()
plt.figure(figsize=(6,4))
sns.barplot(x=gender_counts.index, y=gender_counts.values)
plt.title("Número de hombres y mujeres")
plt.xlabel("Sexo")
plt.ylabel("Cantidad")
plt.show()

# 3.2 Número de personas que sufren o no distintas patologías
diseases = ['Heart_Disease', 'Skin_Cancer', 'Other_Cancer', 'Depression', 'Diabetes', 'Arthritis']
for disease in diseases:
    counts = df[disease].value_counts()
    plt.figure(figsize=(6,4))
    sns.barplot(x=counts.index, y=counts.values)
    plt.title(f"Número de personas con {disease}")
    plt.xlabel(disease)
    plt.ylabel("Cantidad")
    plt.show()

# 3.3 Patologías por sexo
gender_disease = df.groupby('Sex')[diseases].sum()
gender_disease.plot(kind='bar', stacked=True, figsize=(10,6))
plt.title("Patologías por sexo")
plt.xlabel("Sexo")
plt.ylabel("Cantidad de casos")
plt.show()

# 3.4 Matriz de correlación
correlation_matrix = df[diseases].corr()
plt.figure(figsize=(8,6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Matriz de correlación entre variables")
plt.show()

# 3.5 Análisis de tendencia de fumadores y cáncer por rango de edad
# Suponiendo que hay una columna 'Smoking' y 'Age_Group'
age_smoking_cancer = df.groupby(['Age_Group', 'Smoking'])['Skin_Cancer'].sum().unstack()
age_smoking_cancer.plot(kind='bar', stacked=True, figsize=(10,6))
plt.title("Tendencia de fumadores y casos de cáncer por edad")
plt.xlabel("Grupo de edad")
plt.ylabel("Casos de cáncer de piel")
plt.show()

# 3.6 Relación entre el peso y la depresión
sns.boxplot(x=df['Depression'], y=df['BMI'])
plt.title("Relación entre el peso y la depresión")
plt.xlabel("Depresión")
plt.ylabel("Índice de Masa Corporal (BMI)")
plt.show()

# 3.7 Consumo de alimentos por grupos de edad
# Suponiendo que hay columnas relacionadas con el consumo de alimentos
food_columns = ['Fruits_Vegetables', 'Fast_Food', 'Sugary_Drinks']
df.groupby('Age_Group')[food_columns].mean().plot(kind='bar', figsize=(10,6))
plt.title("Consumo de alimentos por grupos de edad")
plt.xlabel("Grupo de edad")
plt.ylabel("Consumo promedio")
plt.show()

# 3.8 Enfermedades a lo largo de los años
# Suponiendo que hay una columna 'Year'
df.groupby('Year')[diseases].sum().plot(figsize=(12,6))
plt.title("Evolución de enfermedades a lo largo de los años")
plt.xlabel("Año")
plt.ylabel("Casos registrados")
plt.legend(loc='upper left')
plt.show()

# 3.9 Consideración del paciente sobre su salud comparado con número de enfermedades
df['Total_Diseases'] = df[diseases].sum(axis=1)
sns.boxplot(x=df['General_Health'], y=df['Total_Diseases'])
plt.title("Percepción de salud vs Total de enfermedades")
plt.xlabel("Percepción de salud")
plt.ylabel("Número de enfermedades")
plt.show()

# 3.10 Consumo de alcohol por grupo de edad
# Suponiendo que hay una columna 'Alcohol_Consumption'
df.groupby('Age_Group')['Alcohol_Consumption'].mean().plot(kind='bar', figsize=(8,5))
plt.title("Consumo de alcohol por grupo de edad")
plt.xlabel("Grupo de edad")
plt.ylabel("Consumo promedio")
plt.show()

# 3.11 Enfermedades del corazón con respecto al consumo de frutas y verduras
sns.boxplot(x=df['Heart_Disease'], y=df['Fruits_Vegetables'])
plt.title("Relación entre enfermedades del corazón y consumo de frutas y verduras")
plt.xlabel("Enfermedad del corazón")
plt.ylabel("Consumo de frutas y verduras")
plt.show()
````

````python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Cargar dataset
df = pd.read_csv("cardiovascular_diseases_risk.csv")

# Separar variables predictoras y variable objetivo
X = df.drop(columns=['Heart_Disease', 'Diabetes', 'Skin_Cancer', 'Other_Cancer', 'Depression', 'Arthritis'])
y_heart = df['Heart_Disease']
y_diabetes = df['Diabetes']

# Dividir en conjunto de entrenamiento y prueba
X_train, X_test, y_train_heart, y_test_heart = train_test_split(X, y_heart, test_size=0.2, random_state=42)
X_train, X_test, y_train_diabetes, y_test_diabetes = train_test_split(X, y_diabetes, test_size=0.2, random_state=42)

# Escalar los datos
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# 4.1 Regresión Logística
log_reg = LogisticRegression()
log_reg.fit(X_train_scaled, y_train_heart)
y_pred_log_reg = log_reg.predict(X_test_scaled)
print("Regresión Logística - Heart Disease")
print(classification_report(y_test_heart, y_pred_log_reg))

# 4.2 Random Forest
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train_heart)
y_pred_rf = rf.predict(X_test)
print("Random Forest - Heart Disease")
print(classification_report(y_test_heart, y_pred_rf))

# 4.3 SVM
svm = SVC(kernel='linear')
svm.fit(X_train_scaled, y_train_heart)
y_pred_svm = svm.predict(X_test_scaled)
print("SVM - Heart Disease")
print(classification_report(y_test_heart, y_pred_svm))

# 4.4 Redes Neuronales
model_nn = keras.Sequential([
    layers.Dense(16, activation='relu', input_shape=(X_train_scaled.shape[1],)),
    layers.Dense(8, activation='relu'),
    layers.Dense(1, activation='sigmoid')
])
model_nn.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model_nn.fit(X_train_scaled, y_train_heart, epochs=20, batch_size=16, validation_split=0.2, verbose=1)
y_pred_nn = (model_nn.predict(X_test_scaled) > 0.5).astype(int)
print("Redes Neuronales - Heart Disease")
print(classification_report(y_test_heart, y_pred_nn))

# 5.1 Validación modelo Redes Neuronales aplicado a 'Heart_Disease'
y_pred_nn_diabetes = (model_nn.predict(X_test_scaled) > 0.5).astype(int)
print("Redes Neuronales - Diabetes")
print(classification_report(y_test_diabetes, y_pred_nn_diabetes))

# 5.3 Validación modelo Random Forest aplicado a 'Skin_Cancer'
y_pred_rf_skin = rf.predict(X_test)
print("Random Forest - Skin Cancer")
print(classification_report(df['Skin_Cancer'][X_test.index], y_pred_rf_skin))
````
