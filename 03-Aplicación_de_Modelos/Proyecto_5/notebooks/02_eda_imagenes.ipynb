# 02 EDA de Imágenes

```python
import os
import random
from PIL import Image
import matplotlib.pyplot as plt
from src.dataset import parse_index_file, image_path

DATA_ROOT = os.path.join('..', 'data', 'dataset-resized')
DOCS = os.path.join('..', 'docs')
train_entries = parse_index_file(os.path.join(DOCS, 'one-indexed-files-notrash_train.txt'))
samples = random.sample(train_entries, 12)
fig, axes = plt.subplots(3, 4, figsize=(10, 8))
for ax, (fname, label) in zip(axes.flat, samples):
    img = Image.open(image_path(DATA_ROOT, fname)).convert('RGB')
    ax.imshow(img)
    ax.set_title(f"{fname} | {label}")
    ax.axis('off')
plt.tight_layout()
```
